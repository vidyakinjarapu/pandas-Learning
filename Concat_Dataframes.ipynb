```python
import pandas as pd
india_weather = pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'banglore'],
    'temperature' : [32, 45, 39],
    'humidity' : [76, 70, 65]
})
india_weather
```

```python
usa_weather = pd.DataFrame({
    'city' : ['newyork', 'da', 'california'],
    'temperature' : [23, 21, 19],
    'humidity' : [56, 50, 45]
})
usa_weather
```

```python
df = pd.concat([india_weather, usa_weather], ignore_index=True) #to get continuous index
```

```python
df
```

```python
df = pd.concat([india_weather, usa_weather], keys=['india', 'us'])
df
```

```python
df.loc['us'] #to access data in specific location
```

```python
temperature_df = pd.DataFrame({
    'city' : ['newyork', 'da', 'california'],
    'temperature' : [23, 21, 19]
}, index=[0, 1, 2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'city' : ['newyork', 'california'],
    'windspeed' : [45, 77]
}, index=[0, 2])
windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df ])
df
```

```python
df = pd.concat([temperature_df,windspeed_df ], axis = 1)
df
```

## Merge in pandas

```python
df1 = pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'hyd'],
    'temperature' : [32, 45, 39]
})
df1
```

```python
df2 = pd.DataFrame({
    'city' : ['mumbai', 'banglore', 'delhi', 'pune'], # if there is an extra city it will ignore while merging the two dataframes
    'humidity' : [67, 76, 59, 45]
})
df2
```

```python
df3 = pd.merge(df1, df2, on="city") # Takes only those cities in common
```

```python
df3
```

```python
df4 = pd.merge(df1, df2, on="city" , how = "outer") #default is "inner" -- intersection -- "left", 'right'
df4
```

```python
df_1 = pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'banglore'],
    'temperature' : [32, 45, 39],
    'humidity' : [76, 70, 65]
})
df_1
```

```python
df_2 = pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'banglore', 'pune'],
    'temperature' : [32, 47, 39, 28],
    'humidity' : [76, 70, 65, 67]
})
df_2
```

```python
df_3 = pd.merge(df_1,df_2, on='city', suffixes=('_left', '_right'), how='outer')
df_3
```
