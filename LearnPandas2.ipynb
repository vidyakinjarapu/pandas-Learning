# Reading and writing CSV file

```python
import pandas as pd
```

```python
df = pd.read_csv("Stock_data.csv", skiprows=1) #to skip one row while loading or header=1
df()
```

```python
df = pd.read_csv("Stock_data.csv")
df
```

```python
#providing a header while running the command
df = pd.read_csv("Stock_data.csv", header=None)
```

```python
df = pd.read_csv("Stock_data.csv", header=None, names=['ticker', 'eps', 'revenue', 'price', 'people'])
df
```

```python
df = pd.read_csv("Stock_data.csv", nrows=3) # loads only 3 rows
df
```

```python
df = pd.read_csv("Stock_data.csv", na_values=["Not Available","n.a."]) # To turn the values into NaN
df
```

```python
df = pd.read_csv("Stock_data.csv", na_values= {
    'eps' : ['Not Available', 'n.a.'],
    'revenue' : ['-1', 'Not Available', 'n.a.'],
    'price' : ['Not Availbale', 'n.a.'],
    'people' : ['Not Available', 'n.a.']
})
df
```

```python
df.to_csv('new.csv', index=False) # To write data to csv file and not save index
```

```python
df.to_csv('new2.csv', index=False, columns=['tickers', 'eps'] ) #To print only 2 columns
```
